#Import Libraries


```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

# Gathering Data for Each Year

## 2022

### Schools List

```python
# Initialize empty lists to store school data from the website
npsn = []
sekolah = []
nilai_total = []
provinsi = []
kota = []
jenis = []

# Base URL for scraping data from the top 1000 schools website
base_url = 'https://top-1000-sekolah.ltmpt.ac.id/?page='

# Loop through each pages of the website
for page in range(1,11) :
  url = base_url + str(page)
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html5lib') # Parse the page content using BeautifulSoup

  awal_page = 1 + 100*(page-1) # Start of the current page (e.g., 1 for page 1, 101 for page 2)
  akhir_page = awal_page+ 99 # End of the current page (e.g., 100 for page 1, 200 for page 2)

  # Loop through each row (representing a school) on the current page
  for i in range(awal_page,akhir_page+1) :
    table_row = soup.find('tr',{'data-key' : i})
    td_npsn = table_row.find_all('td')[2]
    td_sekolah = table_row.find_all('td')[3]
    td_nilai_total = table_row.find_all('td')[4]
    td_provinsi = table_row.find_all('td')[5]
    td_kota = table_row.find_all('td')[6]
    td_jenis = table_row.find_all('td')[7]
    npsn.append(td_npsn.text)
    sekolah.append(td_sekolah.text)
    nilai_total.append(td_nilai_total.text)
    provinsi.append(td_provinsi.text)
    kota.append(td_kota.text)
    jenis.append(td_jenis.text)
```

### Sub-test

```python
# Initialize empty lists to store subtest scores for each school
pk = []
pbm = []
pu = []
ppu = []

# Base URL for retrieving detailed subtest scores from the website
base_url_subtes = 'https://top-1000-sekolah.ltmpt.ac.id/site/page2022?id='

# Loop through the first 1000 schools (IDs range from 1 to 1000)
for i in range(1,1001) :
  url = base_url_subtes + str(i)
  r = requests.get(url)
  soup_nilai = BeautifulSoup(r.content, 'html5lib')
  nilai_pk = soup_nilai.find_all('b')[6].text #Nilai PK
  nilai_pbm = soup_nilai.find_all('b')[8].text #Nilai PBM
  nilai_pu = soup_nilai.find_all('b')[10].text #Nilai PU
  nilai_ppu = soup_nilai.find_all('b')[12].text #Nilai PPU
  pk.append(nilai_pk)
  pbm.append(nilai_pbm)
  pu.append(nilai_pu)
  ppu.append(nilai_ppu)
```

### Create DataFrame

```python
rank = list(range(1,1001))
tabel_2022 = pd.DataFrame(list(zip(rank, npsn, sekolah, provinsi, kota, jenis, nilai_total, pk, pbm, pu, ppu)), columns = ['Peringkat','NPSN', 'Nama Sekolah', 'Provinsi', 'Kota/Kabupaten', 'Jenis', 'Nilai Total', 'PK', 'PBM', 'PU', 'PPU'])
tabel_2022.head(5)
```
